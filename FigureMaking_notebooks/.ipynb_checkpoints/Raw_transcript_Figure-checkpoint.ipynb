```python
# load codebook
import pandas as pd
codebook = pd.read_csv(r'/mnt/Y/Lab/Raw_custom_MERFISH_Imaging_data/20250827_3924_GP/analysis_notebooks/Final_codebook_BG_673gn_30_bits.csv',index_col = 0)
DRD1_barcode = codebook.loc['DRD1'].values[:-1]
DRD2_barcode = codebook.loc['DRD2'].values[:-1]
```

```python
DRD1_onbits = np.where(DRD1_barcode)
DRD2_onbits = np.where(DRD2_barcode)
onbits = np.unique(np.hstack([DRD1_onbits,DRD2_onbits]))
channels = ['750','cy5','cy3']
used_channels = [channels[b%3] for b in onbits]
used_channels
```

```python
onbits
```

```python
hyb_rounds = [(b-1)//3 + 1 for b in onbits]
hyb_rounds
```

- Only read the raw file where on-bit show up.

```python
from dask import array as da
DAPI = []
ch750 = []
chcy5 = []
chcy3 = []

# load in the direct correction file for the FOV 
drift_corrrection = np.load(r'/mnt/Y/Lab/Raw_custom_MERFISH_Imaging_data/20250511_XXhBG_1311' + os.sep + r'MERmake_decoding' + os.sep + f'driftNew_Conv_zscan__{selected_FOV:04d}--.pkl', allow_pickle=True)[0]

img_corrected = {}
for hyb_idx in np.unique(hyb_rounds):
    img = da.from_zarr(r'/mnt/Y/Lab/Raw_custom_MERFISH_Imaging_data/20250511_XXhBG_1311' + os.sep +  f'H{hyb_idx}_MER_set1' + os.sep + f'{selected_FOV:04d}' + os.sep + 'data')[1:]
    drift = drift_corrrection[hyb_idx - 1][0]
    for ch_idx in range(4):
        # apply drift correction 
        img_ = img[ch_idx::4]
        img_corrected[f'Ch_{ch_idx}'] = img_corrected.get(f'Ch_{ch_idx}',[]) + [da.roll(img_, axis = [0,1,2], shift = np.round(drift).astype(int))]

```

```python
# take the maximum projection
for k in img_corrected:
    img_corrected[k] = da.stack(img_corrected[k]).max(axis = 0)
```

- How to identifiy the transcript pacta. so that we can show the real transcript.
- Procedure. 
- identify the channel and the round at which the a gene has on-bits, 
- deconvolve those image, 
- threhsold and take the mean of all those images. And this should separate the pancta of that transcript.
